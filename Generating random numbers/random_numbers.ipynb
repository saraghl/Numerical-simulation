```python
#### 2
# c= 1.00248

def generator2(N):
  random_list=[]
  for i in range(N):
    t=np.random.random()
    x=-np.log(1-t*(1-np.exp(-6)))
    random_list.append(x)
  return random_list

def generator2v2(N):
    t=np.random.rand(N)
    x=-np.log(1-t*(1-np.exp(-6)))
    return list(x)
```

# **Integrate**

```python
# Monte-Carlo Method
def f(x):
    return -(x**2)+1
#lambda ?
def integrate(N,f):
    k=0
    for i in range(N):
        x=2*(np.random.rand()-0.5)
        y=np.random.rand()
        if y<=f(x):
            k=k+1
    return 2*k/N
```

```python
integrate(1000000,f)
```

```python
#small squares
def integratev2(dx,f):
    sum=0
    for i in np.arange(-1,1,dx):
        s=f(i)*dx
        sum=sum+s
    return sum
```

```python
integratev2(0.01,lambda x:-(x**2)+1)
```
